```python
from sentence_transformers import SentenceTransformer

#Using paraphrase because my computer is SLOW
model = SentenceTransformer('paraphrase-MiniLM-L3-v2')
```

```python
list(df_clusternames)
```

```python
occupation_embeddings = model.encode(df_wages['Occupation'].tolist())
cluster_embeddings = model.encode(df_clusternames['description'].tolist())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

#the cosine similarity between occupation embeddings and cluster embeddings
similarities = cosine_similarity(occupation_embeddings, cluster_embeddings)

#the index of the max similarity (so we can see whats the best match!)
closest_clusters_indices = similarities.argmax(axis=1)

#mapping to codes and names
# df_wages['Career Cluster Code'] = closest_clusters_indices
df_wages['Career Cluster Name'] = [df_clusternames.iloc[index]['description'] for index in closest_clusters_indices]
```
